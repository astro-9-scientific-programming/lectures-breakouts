```python
def derivativeL(func,x,h=0.01):
    return (func(x+h)-func(x))/h

def derivativeC(func,x,h=0.01):
    return (func(x+h/2)-func(x-h/2))/h

def derivativeR(func,x,h=0.01):
    return (func(x)-func(x-h))/h
```

```python
h = np.logspace(-16,1,100)
```

```python
derivL = derivativeL(func,np.pi/3,h)
derivC = derivativeC(func,np.pi/3,h)
derivR = derivativeR(func,np.pi/3,h)
```

```python
def integral(func, x_0 = 0, x_1 = 1, N=100): #functional 
    x = np.linspace(x_0,x_1,N)
    dx = x[1]-x[0]
    integral = np.sum(func(x))*dx
    return integral
```

```python
def deriv(func,x,h=0.0001):
    return (func(x+h)-func(x))/h


x_0 = 2.5
fx_0 = func(x_0)
fp_0 = deriv(func,x_0)
```

```python
plt.plot(x,fx,"b")
plt.plot(x_0,fx_0,"o",markersize=20)
plt.plot([x_0-0.5,x_0+0.5],fp_0*np.array([x_0-0.5,x_0+0.5])-0.335)
```

```python
x_1 = x_0 - 1*fp_0 #updated position

fx_1 = func(x_1)


```

```python
plt.plot(x,fx,"b")
plt.plot(x_0,fx_0,"o",markersize=10)

plt.plot(x_1,fx_1,"o",markersize=10)
```

```python
plt.plot(x,fx,"b")

x = 2.5
fx = func(x)
fp = deriv(func,x)

fp_L = fp
x_L = x
while abs(fp) > 0.00001: #convergence criteria or tolerance
    x = x_L - fp_L
    fx = func(x)
    fp = deriv(func,x)
    print(x,fx)
    fp_L = fp
    x_L = x
    
    plt.plot(x,fx,"o",color="black",markersize=10)
    
```
